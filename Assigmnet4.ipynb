```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv"
df = pd.read_csv(url)
# Calculate correlation matrix
correlation = df.corr()

# Show correlation of each feature with the target column 'Class'
corr_with_target = correlation['Class'].sort_values(ascending=False)
print(corr_with_target)
plt.figure(figsize=(8,6))
corr_with_target[:-1].sort_values(ascending=False).plot(kind='bar', color='teal')
plt.title('Correlation of Each Feature with Target Variable (Class)')
plt.ylabel('Correlation Coefficient')
plt.xlabel('Features')
plt.show()
```

Correlation Analysis Report

Objective:
To identify which features are most related to the target variable Class (Fraud = 1, Non-Fraud = 0).

Findings:

After performing correlation analysis, we found that:

V17, V14, and V12 are the top 3 features most related to the target variable.

These features show strong negative correlations, indicating that lower values of these components are more likely associated with fraud.

Other moderately related features include V10, V16, and V18, but they are slightly less influential.

Insight:
These 3 key features (V17, V14, V12) will play an important role in model training, as they carry the most predictive information for detecting fraudulent transactions.
